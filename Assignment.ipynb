```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the datasets
User_details=pd.read_excel("/content/User_details.xlsx")
Cooking_sessions=pd.read_excel("/content/Cooking_sessions.xlsx")
Order_details=pd.read_excel("/content/Order_details.xlsx")
```

```python
merged_data = Order_details.merge(Cooking_sessions, on='Session ID', suffixes=('_order', '_session'))
```

```python
# Rename 'User ID_order' to 'User ID' for merging with User_details
merged_data.rename(columns={'User ID_order': 'User ID'}, inplace=True)

# Merge with User_details
final_data = merged_data.merge(User_details, on='User ID')
```

```python
final_data.head()
```

```python
# Drop the redudant columns
final_data.drop(columns=['Phone','Email','Session Start','Session End'], inplace=True)
```

```python
# Checking null values
final_data.isnull().sum()
```

```python
final_data['Rating'] = final_data['Rating'].fillna(final_data['Session Rating'])
final_data.dropna(inplace=True)
```

```python
final_data.isnull().sum()
```

```python
final_data.dtypes
```

```python
# Relationship between cooking sessions and user orders
session_order_corr = final_data['Session Rating'].corr(final_data['Rating'])
print(f"Correlation between session rating and order rating: {session_order_corr:.2f}")
```

```python
# Identifying popular dishes by frequency and revenue
popular_dishes = final_data.groupby('Dish Name_session').agg(
    Total_Orders=('Order ID', 'count'),
    Total_Revenue=('Amount (USD)', 'sum')
).sort_values(by='Total_Orders', ascending=False)
print("Top 5 Popular Dishes by Frequency and Revenue:")
popular_dishes.head()
```

```python
# Average rating by age group
final_data['Age Group'] = pd.cut(final_data['Age'], bins=[20, 30, 40, 50], labels=['20-30', '30-40', '40-50'])
avg_rating_by_age = final_data.groupby('Age Group')['Rating'].mean()
print("Average Rating by Age Group:")
avg_rating_by_age
```

```python
# Meal preferences by location
meal_by_location = final_data.groupby(['Location', 'Meal Type_order']).size().unstack(fill_value=0)

print("Meal Preferences by Location:")
meal_by_location
```

```python
#Popular dishes visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=popular_dishes.index[:5], y=popular_dishes['Total_Orders'][:5], palette='viridis')
plt.title('Top 5 Dishes by Order Frequency')
plt.xlabel('Dish Name')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
# Ratings by Age Group
plt.figure(figsize=(8, 5))
avg_rating_by_age.plot(kind='bar', color=['skyblue','lightgreen','coral'])
plt.title('Average Rating by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Average rating by Dish
avg_rating_by_dish =final_data.groupby('Dish Name_order')['Rating'].mean()
plt.figure(figsize=(10,4))
avg_rating_by_dish.plot(kind='bar',color=['skyblue','coral','mediumseagreen','gold','orchid'])
plt.title('Average Ratings by dishes')
plt.xlabel('Dish name')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Duartion by Dishes
plt.figure(figsize=(12,6))
sns.boxplot(data=final_data, x='Dish Name_session',y='Duration (mins)',palette='Set2')
plt.title('Cooking Duration by Dish')
plt.xlabel('Dish Name')
plt.ylabel('Duration (mins)')
plt.xticks(rotation=45)
plt.show()
```

```python
# mean cooking Duration by user
total_duration_by_user = final_data.groupby('User Name')['Duration (mins)'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
total_duration_by_user.plot(kind='bar',color='skyblue')
plt.title('Mean cooking Duration by user')
plt.xlabel('User Name')
plt.ylabel(' Mean Duration (minutes) ')
plt.show()
```

```python
## Count the number of occurences for each Meal Type
meal_type_counts = final_data['Meal Type_session'].value_counts()

plt.figure(figsize=(8,5))
meal_type_counts.plot(kind='bar', color=['mediumseagreen','gold','skyblue'])
plt.title("Meal Type Distribution")
plt.xlabel("Meal Type")
plt.ylabel("Number of sessions")
```

```python
# Distribution of favorite Meals
fav_meal_counts=final_data['Favorite Meal'].value_counts()
plt.figure(figsize=(8,5))
fav_meal_counts.plot(kind='bar',color='teal')
plt.title("Distribution of favorite meal")
plt.xlabel("Meal type")
plt.ylabel("Number of users")
plt.xticks(rotation=0)
plt.show()
```

```python
# Total revenue by Dish
revenue_by_meal_type = final_data.groupby('Dish Name_session')['Amount (USD)'].sum()
revenue_by_meal_type= revenue_by_meal_type.sort_values(ascending=False)
plt.figure(figsize=(8,6))
revenue_by_meal_type.plot(kind='bar',color=["#FF9999","#66B2FF","#99FF99"])
plt
```

```python
# Proportion of Orders by Location (Pie Chart)
location_analysis = final_data.groupby('Location').agg({
    'Total Orders': 'sum',
    'Amount (USD)': 'mean',
    'Duration (mins)': 'mean',
    'Rating': 'mean'
}).reset_index()
plt.figure(figsize=(6, 6))
plt.pie(
    location_analysis['Total Orders'],
    labels=location_analysis['Location'],
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title('Proportion of Total Orders by Location')
plt.show()
```

```python
order_status = final_data['Order Status'].value_counts()

plt.figure(figsize=(4, 4))
order_status.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'salmon'], labels=order_status.index)
plt.title('Order Status Distribution')
plt.ylabel('')
plt.show()
```
